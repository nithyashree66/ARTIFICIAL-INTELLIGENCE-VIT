**TASK 1 - CREATE A PANDAS DATAFRAME WITH NUMPY RANDOM VALUES**

```python
data_random = np.random.rand(4,4)
```

```python
data_random
```

**TASK 2 - RENAME THE TASK-1 DATAFRAME COLUMN NAMES TO RANDOM VALUE 1,RANDOM VALUE 2,RANDOM VALUE 3,RANDOM VALUE4**

```python
df = pd.DataFrame(data_random, columns=['Random value 1',
                  'Random value 2','Random value 3','Random value 4'])
```

**TASK 5 - DISPLAY THE RANDOM VALUE 2 &RANDOM VALUE 3 COLUMNS WITH LOCATION AND INDEX LOCATION METHOD**

the location (loc) uses labels or index values that specifies the rows with indeces.

```python
#need to specify the labels
print(df.loc[:,['Random value 2','Random value 3']])
```

```python
#need to specify the column index
print(df.iloc[:,[1,2]])
```
