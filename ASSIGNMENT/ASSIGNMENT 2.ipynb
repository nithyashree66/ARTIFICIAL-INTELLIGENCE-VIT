**BUILD AN ANN MODEL FOR DRUG CLASSIFICATION**

**TASK 1 - IMPORT THE DATASET AND DO DATA PREPROCESSING**

import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read the CSV file

```python
df = pd.read_csv('/content/drug200.csv')
```

```python
df
```

Get a statistical analysis of the data

```python
df.describe()
```

check for null values

```python
df.isnull().sum()
```

seperate features and labels

```python
features = df.drop('Drug',axis=1)
labels = df[['Drug']]
```

convert categorical data to numerical data

```python
#one hot encoding
numerical_data = pd.get_dummies(features)
```

split the data into training and testing sets

```python
normalized_features = (numerical_data - numerical_data.mean()) / numerical_data.std()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(normalized_features, labels, test_size=0.2, random_state=42)
```

```python
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

**TASK 2 - BUILD AN ANN MODEL**

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

input layer

```python
model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
```

hidden layer

```python
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
```

output layer

```python
model.add(Dense(units=5,activation='softmax'))
```

compile the model

```python
model.compile(loss='categorical_crossentropy',optimizer = 'adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train,y_train,batch_size=32,epochs = 100)
```

**TASK 3 - TEST THE MODEL WITH RANDOM DATA**

```python
import numpy as np
```

```python
import numpy as np

num_samples = 5
random_features = np.random.randn(num_samples, X_train.shape[1])

predictions = model.predict(random_features)

predicted_drugs = [labels[np.argmax(pred)] for pred in predictions]

print(predicted_drugs)
```
